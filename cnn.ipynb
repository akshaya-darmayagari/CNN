```python
# Import the necessary libraries 
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image  import ImageDataGenerator
```

```python
#Import the dataset 
dataset_path=r"C:\Users\aksha\Desktop\AI&DS TRAINING\CNN\PetImages"
```

```python
# Image prprocessing
datagen=ImageDataGenerator(
    #Normalize pixels 
    rescale=1./255,
    # Data augementation
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,

    # Split dataset 
    validation_split=0.2 # Validation split used to create train and validation data
)
```

```python
# Load traing data 
# Training data 
train_Data=datagen.flow_from_directory(
    dataset_path,
    target_size=(128,128),
    batch_size=32,
    class_mode='binary',
    subset='training'
)
```

```python
# load validation data 
val_data=datagen.flow_from_directory(
    dataset_path,
    target_size=(128,128),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
print(train_Data.class_indices)
```

```python
# Building CNN model 
model=keras.Sequential()
```

```python
# 1.Simple Images ---> Few Layers 
# ex: handwritten numbers
# (28*28)
# 2 conc 2 pool
 
# 2.Medium Complexity
# Cats vs Dogs
# (128*128)
# (224*224)
# 3 Conv + 3 Pool.
```

```python
# First convolutional layer 
model.add(
    keras.layers.Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation='relu',
        input_shape=(128,128,3)
    )
)
```

```python
# First pooling layer 
model.add(
    keras.layers.MaxPooling2D(
        pool_size=(2,2)
        )
)
```

```python
# 2 convolutional layer 
model.add(
    keras.layers.Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation='relu'
    )
)
```

```python
# Second pooling layer 
model.add(
    keras.layers.MaxPooling2D(
        pool_size=(2,2)
    )
)
```

```python
# Flatten 
model.add(keras.layers.Flatten())
```

```python
# Dense layer 
model.add(
    keras.layers.Dense(
        units=128,
        activation='relu'
        )
)
```

```python
# Output layer 
model.add(
    keras.layers.Dense(
        units=1,
        activation='sigmoid'
    )
)
```

```python
# Compile the model 
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
# Train 
history=model.fit(
    train_Data,
    epochs=2,
    validation_data=val_data
)
```

```python
# Evaluate the model 
loss,accuracy=model.evaluate(val_data)
print("Accuracy: ",accuracy)
```
